```python
import os
import scipy.io as scio
import numpy as np
# get original data



def GetOrigin(n,x):#n is the ratio of train to test, xyz are the dimensions of data
    data_train = np.empty([1250, 10, 10, 16])
    data_test = np.empty([255, 10, 10, 16])
    data_train = scio.loadmat('data_1250+255.mat')['data_train']
    data_test = scio.loadmat('data_1250+255.mat')['data_test']
    # change ratio of train to test
    temNum = 1505
    allSpace = int(temNum/5)
    Temp = temNum-5
    TraNum = (Temp//(n+1))*n+5
    TesNum = temNum - TraNum
    all_train = np.empty([temNum, 10, 10, 16])
    for i in range(0,5):
        tem_tra = data_train[250*i:250*i+250,:,:,:]
        tem_tes = data_test[51*i:51*i+51,:,:,:]
        all_train[301*i:301*i+250,:,:,:] = tem_tra
        all_train[301*i+250:301*i+301,:,:,:] = tem_tes
    afRatioTrain = np.empty([TraNum, 10, 10, 16])
    afRatioTest = np.empty([TesNum, 10, 10, 16])
    trainSpace = int(TraNum/5)
    testSpace = int(TesNum/5)
    for i in range(0,5):
        tem = all_train[allSpace*i:allSpace*i+trainSpace,:,:,:]
        left = all_train[allSpace*i+trainSpace:allSpace*i+allSpace,:,:,:]
        afRatioTrain[trainSpace*i:trainSpace*i+trainSpace,:,:,:] = tem
        afRatioTest[testSpace*i:testSpace*i+testSpace,:,:,:] = left
    # change the dimension
    temTrain = afRatioTrain.reshape(TraNum, -1)
    temTest = afRatioTest.reshape(TesNum, -1)
    if x[0]*x[1]*x[2] == 1600:
        final_train = temTrain.reshape(TraNum, x[0], x[1], x[2])
        final_test = temTest.reshape(TesNum, x[0], x[1], x[2])
    # label the data
    label_trian = np.empty([TraNum])
    label_test = np.empty([TesNum])
    for i in range(0, 5):
        label_trian[trainSpace*i:trainSpace*i+trainSpace] = i
        label_test[testSpace*i:testSpace*i+testSpace] = i
    return final_train,final_test,label_trian,label_test


def GetSpiral(inputShape):#n is the ratio of train to test, xyz are the dimensions of data
    data_train = np.empty([1206, 8, 10, 1])
    data_test = np.empty([299, 8, 10, 1])

    label_trian = np.empty([1206])
    label_test = np.empty([299])
    label_trian[0:241] =0
    label_trian[241:482] =1
    label_trian[482:723] =2
    label_trian[723:962] =3
    label_trian[962:1206] =4
    label_test[0:60] =0
    label_test[60:120] =1
    label_test[120:180] =2
    label_test[180:239] =3
    label_test[239:299] =4


    data_train = scio.loadmat('SpiralDataset.mat')['train']
    data_test = scio.loadmat('SpiralDataset.mat')['test']
    return data_train,data_test,label_trian,label_test

def GetNew(inputShape,DsetName):#n is the ratio of train to test, xyz are the dimensions of data
    data_train = scio.loadmat('.//datasets//'+DsetName+'//Dataset_train.mat')['TrainfinalDataset']
    data_test = scio.loadmat('.//datasets//'+DsetName+'//Dataset_test.mat')['TestfinalDataset']
    label_trian = scio.loadmat('.//datasets//'+DsetName+'//Trainlabel.mat')['Trainlabel']
    label_test = scio.loadmat('.//datasets//'+DsetName+'//Testlabel.mat')['Testlabel']
    return data_train,data_test,label_trian,label_test
```
